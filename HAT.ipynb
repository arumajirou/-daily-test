```python
import os
import shutil

# ディレクトリの設定
gt_directory = os.environ.get('GT_DIRECTORY', '/content/HAT/datasets/Set5/GTmod4/')
lq_directory = os.environ.get('LQ_DIRECTORY', '/content/HAT/datasets/Set5/LRbicx4/')

# GTディレクトリからLQディレクトリにファイルをコピーし、名前を変更
for filename in os.listdir(gt_directory):
    if filename.endswith('.png'):
        new_filename = filename.replace('.png', 'x4.png')
        source_path = os.path.join(gt_directory, filename)
        destination_path = os.path.join(lq_directory, new_filename)

        if not os.path.exists(destination_path):
            shutil.copy(source_path, destination_path)
            print(f'Copied and renamed: {source_path} to {destination_path}')
        else:
            print(f'File already exists: {destination_path}')

print("File copying and renaming completed.")
```

```python
import os
import shutil

def create_and_rename_files(src_directory, dest_directory, old_suffix, new_suffix):
    # Create destination directory if it doesn't exist
    if not os.path.exists(dest_directory):
        os.makedirs(dest_directory)
        print(f"Directory created: {dest_directory}")

    # Move and rename files from source to destination directory
    for filename in os.listdir(src_directory):
        if filename.endswith(old_suffix):
            new_filename = filename.replace(old_suffix, new_suffix)
            old_path = os.path.join(src_directory, filename)
            new_path = os.path.join(dest_directory, new_filename)
            shutil.move(old_path, new_path)
            print(f'Moved and renamed: {old_path} to {new_path}')

# Set the source and destination directories and suffixes
src_directory_x2 = '/content/HAT/datasets/Set5/LRbicx4/'
dest_directory_x2 = '/content/HAT/datasets/Set5/LRbicx2/'
src_directory_x3 = '/content/HAT/datasets/Set5/LRbicx4/'
dest_directory_x3 = '/content/HAT/datasets/Set5/LRbicx3/'

# Move and rename files in the x2 directory
create_and_rename_files(src_directory_x2, dest_directory_x2, 'x2.png', '*x2.png')

# Move and rename files in the x3 directory
create_and_rename_files(src_directory_x3, dest_directory_x3, 'x3.png', '*x3.png')
```
