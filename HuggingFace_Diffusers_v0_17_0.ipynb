- A100とは、最新のタイプのGPUです。性能は非常に高くて機能も非常に多いです。有料版で使えます。

- 目的が難しくて予算が多いなら、有料版のP100やV100やA100がおすすめかもしれません。

## UniDiffuserとは何か
UniDiffuser（ユニディフューザー）とは、テキスト（文字や言葉）や画像（絵や写真）などのデータを扱うことができる人工知能（コンピューターが自分で考えたり学習したりする技術）の一種です¹²³。
UniDiffuserは、diffusion（拡散）という方法を使って、データを生成（作り出す）したり、変換（別の形に変える）したりします¹²³。
diffusionとは、データにノイズ（雑音やごちゃごちゃしたもの）を加えたり減らしたりすることで、データの品質（きれいさや詳しさ）や内容（何を表しているか）を変えることです¹²³。
UniDiffuserは、一つのモデル（人工知能の仕組みや知識）で、様々なデータを扱うことができます¹²³。
モデルとは、人工知能が学習したパラメータ（数値や重み付け）や構造（計算の仕方や特徴）のことです¹²³。
パラメータとは、人工知能が学習した知識や重み付けを表す数値です¹²³。
構造とは、人工知能が計算する仕方や特徴を表すものです¹²³。

## UniDiffuserができること
UniDiffuserは、以下のようなことができます¹²³。

- 無条件の画像とテキストの生成
  - 何も入力しなくても、画像やテキストを自動的に作ることができます。
- 共同画像テキスト生成
  - 画像とテキストの両方を入力すると、それらに関連した画像やテキストを作ることができます。
- テキストからの画像生成
  - テキストを入力すると、そのテキストに合った画像を作ることができます。
- 画像からのテキスト生成
  - 画像を入力すると、その画像に合ったテキストを作ることができます。
- 画像のバリエーション
  - 画像を入力すると、その画像に似ているけど少し違う画像を作ることができます。
- テキストのバリエーション
  - テキストを入力すると、そのテキストに似ているけど少し違うテキストを作ることができます。

## UniDiffuserの仕組み
UniDiffuserは、以下のような仕組みで動いています¹²³。

- データを扱うときには、まずデータにノイズを加えて、品質や内容を変えます。
- ノイズを加えるときには、timestep（時間の区切り）というものを使って、どれだけノイズを加えるかを決めます。
- timestepとは、時間の区切りのことで、0から始まって1に近づくにつれて、ノイズが増えていきます。
- 例えば、timestepが0のときは、ノイズがなくて元のデータと同じです。timestepが1に近いときは、ノイズがたくさんあって元のデータと違います。
- データにノイズを加えたら、次にノイズを減らして、元のデータに近づけます。
- ノイズを減らすときには、人工知能が学習したパラメータや構造を使って、ノイズの値を予測（推測）します。
- 予測とは、人工知能が知っている情報から、知らない情報を推測することです。
- 予測したノイズの値をデータから引くことで、ノイズを減らします。
- ノイズを減らすときにもtimestepを使って、どれだけノイズを減らすかを決めます。
- 例えば、timestepが1に近いときは、ノイズがたくさんあって元のデータと違うので、たくさんノイズを減らします。timestepが0のときは、ノイズがなくて元のデータと同じなので、もうノイズを減らしません。
- データにノイズを加えたり減らしたりすることで、データの品質や内容を変えることができます。

ControlNetとは何か

ControlNet（コントロールネット）とは、diffusion（拡散）という方法を使って、画像を生成（作り出す）したり、変換（別の形に変える）したりする人工知能（コンピューターが自分で考えたり学習したりする技術）の一種です³⁴⁵。 ControlNetは、extra conditions（追加の条件）というものを使って、diffusion models（拡散モデル）を制御（操作や調整）することができます³⁴⁵。 extra conditionsとは、画像の内容やスタイルなどを指定するテキストや画像などのデータです³⁴⁵。 diffusion modelsとは、diffusionという方法で画像を生成したり変換したりする人工知能のモデルです³⁴⁵。 diffusionとは、画像にノイズ（雑音やごちゃごちゃしたもの）を加えたり減らしたりすることで、画像の品質（きれいさや詳しさ）や内容（何を表しているか）を変えることです³⁴⁵。

ControlNetができること

ControlNetは、以下のようなことができます³⁴⁵。

画像から画像への変換
画像を入力すると、その画像に関連した別の画像を作ることができます。
例えば、人物の顔写真を入力すると、その人物の別の表情や髪型やメイクなどを作ることができます。
画像からテキストへの変換
画像を入力すると、その画像に合ったテキストを作ることができます。
例えば、風景写真を入力すると、その風景について説明したり感想を述べたりする文章を作ることができます。
テキストから画像への変換
テキストを入力すると、そのテキストに合った画像を作ることができます。
例えば、「赤い花が咲く森」というテキストを入力すると、そのテキストに忠実な画像を作ることができます。
画像の修復
画像に欠損（消えたり壊れたりした部分）やノイズがある場合、それらを修復（元通りに直す）することができます。
例えば、古い写真にあるシミや傷などを消して、鮮明な写真にすることができます。
ControlNetの仕組み

ControlNetは、以下のような仕組みで動いています³⁴⁵。

データを扱うときには、まずデータにノイズを加えて、品質や内容を変えます。
ノイズを加えるときには、timestep（時間の区切り）というものを使って、どれだけノイズを加えるかを決めます。
timestepとは、時間の区切りのことで、0から始まって1に近づくにつれて、ノイズが増えていきます。
例えば、timestepが0のときは、ノイズがなくて元のデータと同じです。timestepが1に近いときは、ノイズがたくさんあって元のデータと違います。
データにノイズを加えたら、次にノイズを減らして、元のデータに近づけます。
ノイズを減らすときには、人工知能が学習したパラメータや構造を使って、ノイズの値を予測（推測）します。
予測とは、人工知能が知っている情報から、知らない情報を推測することです。
予測したノイズの値をデータから引くことで、ノイズを減らします。
ノイズを減らすときにもtimestepを使って、どれだけノイズを減らすかを決めます。
例えば、timestepが1に近いときは、ノイズがたくさんあって元のデータと違うので、たくさんノイズを減らします。timestepが0のときは、ノイズがなくて元のデータと同じなので、もうノイズを減らしません。
データにノイズを加えたり減らしたりすることで、データの品質や内容を変えることができます。
ControlNetは、extra conditions（追加の条件）というものを使って、diffusion models（拡散モデル）を制御（操作や調整）します。
extra conditionsとは、画像の内容やスタイルなどを指定するテキストや画像などのデータです。
ControlNetは、diffusion models（拡散モデル）のパラメータや構造をコピー（複製）して、locked copy（ロックされたコピー）とtrainable copy（訓練可能なコピー）という二つのものに分けます。
locked copyとは、元のパラメータや構造をそのまま保持して変更しないものです。
trainable copyとは、extra conditionsに合わせてパラメータや構造を変更できるものです。
ControlNetは、locked copyとtrainable copyの両方を使って、データにノイズを加えたり減らしたりします。
locked copyは、元のデータに忠実な画像を生成するために使われます。
trainable copyは、extra conditionsに忠実な画像を生成するために使われます。
ControlNetは、locked copyとtrainable copyの出力を混ぜ合わせて（重み付けして）、最終的な画像を生成します。
混ぜ合わせる際には、timestepやextra conditionsなどに応じて重み付けを変えます。
例えば、timestepが0のときは、locked copyの出力に重み付けして元のデータに近い画像を生成します。timestepが1に近いときは、trainable copyの出力に重み付けしてextra conditionsに近い画像を生成します。
