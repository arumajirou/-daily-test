# 異常検知、因果推論、時系列分析

- 異常検知とは、期待される動作から逸脱したデータの異常なパターンを特定する作業である。異常検知は、不正、障害、エラーなどの検知に利用できる。
- 因果関係推論とは、観測データや実験データから変数間の因果関係を推論する作業である。因果関係の推論は、介入、政策、治療などの効果を理解するために使用することができる。
- 時系列分析とは、株価、天候、センサーの測定値など、時間の経過とともに変化するデータを研究することである。時系列分析は、予測、異常検知、分類などに利用することができる。
- 異常検知、因果推論、時系列分析を組み合わせることで、時系列データにおける異常の原因と影響を調べることができる。例えば、異常検知は時系列データの異常値や変化を特定するために[^1^][1]、因果推論は異常の影響や帰属を推定するために[^2^][2]、時系列分析は異常の挙動をモデル化して予測するために使用することができる。

## モデル

- 異常検知の因果推論と時系列解析を組み合わせて高い予測精度を持つモデルには、以下のようなものがある。
  - ベイズ型構造時系列（BSTS）。BSTS（Bayesian Structural Time Series）：ベイズのフレームワークを用いて、時系列のトレンド・季節性・回帰などの構造要素を推定するモデルである。回帰係数に対するスパイク・アンド・スラブ事前分布を用いた異常検出[^3^][3]、合成制御法を用いた因果関係推定[^4^][4]、状態空間モデルを用いた時系列解析[^5^][5]が可能である。
  - DeepAR：DeepARは、RNN（Recurrent Neural Network）を用いて、時系列の自己回帰ダイナミクスを学習するモデルです。DeepARは、確率予測や閾値処理を用いた異常検知[^6^][6]、反実仮想推論を用いた因果推論、注目メカニズムを用いた時系列解析などを行うことができます。

# 多変量・多段・時系列解析

- 多変量時系列解析とは、株価、気象パターン、センサー読み取り値など、時間と共に変化する複数の変数からなるデータを研究するものである。多変量時系列解析は、予測、異常検知、分類などに利用することができる。
- 多段階時系列分析とは、カスタマージャーニー、生産プロセス、病気の進行など、時間とともに変化する複数の段階やレベルで構成されるデータを研究するものである。多段階時系列解析は、セグメンテーション、クラスタリング、最適化などに利用することができる。
- 多変量解析と多段階時系列解析は、時系列データの複雑で階層的な構造と挙動を研究するために組み合わせることができる。例えば、多変量・多段時系列解析を使って、異なるチャネルとの相互作用に基づいて顧客の解約をモデル化して予測したり[^1^][1]、需要在庫品質などの複数の要因に基づいて生産スケジュールを最適化したり[^2^][2]、複数のバイオマーカーに基づいて病気を診断して治療する[^3^][3]などが可能である。

## モデル

- 多変量解析と多段階時系列解析を組み合わせて高い予測精度を持つモデルには、以下のようなものがある。
  - ベクトル自己回帰移動平均(VARMA)。VARMAは、単一変数の代わりに変数のベクトルを使用することにより、一変量ARMAモデルを多変量設定に拡張したモデルです。VARMAは、時系列における複数の変数間の線形従属性とダイナミクスを捉えることができます。VARMAは、マクロ経済変数[^4^][4]、電力価格[^5^][5]などの予測に利用されている。
  - ベイジアンネットワーク（BN）。BN は、変数の集合とその条件依存性を有向無サイクルグラフで表現するモデルである。BNは、時系列における複数の変数間の確率的関係や不確実性を捉えることができる。BNは交通流の予測[^6^][6]、株価の予測などに利用されている。

# 時系列予測の動向と将来展望

時系列予測は、一連の時間を通じて事象を予測する手法である。将来のトレンドが過去のトレンドと同様に推移するという前提で、過去のトレンドを分析することで将来の事象を予測する[^1^][1][^2^][6]。天文学、経済学、金融、マーケティング、医学、気象学など、さまざまな分野で応用されている[^1^][1][^3^][4]。

## 研究動向

時系列予測に関する現在の研究動向としては、以下のようなものがある。

- **ディープラーニング**ディープラーニングは、複雑で高次元の時系列データに対して有望な結果を示している。ディープラーニングは、複数の時間スケールやモダリティにまたがる非線形パターンや依存関係を捉えることができる。時系列予測のための深層学習アーキテクチャの例としては、リカレントニューラルネットワーク（RNN）、畳み込みニューラルネットワーク（CNN）、注意メカニズム、トランスフォーマー、グラフニューラルネットワーク（GNN）[^4^][8]がある。
- **確率的予測**:確率論的予測は、将来の事象について、点推定だけでなく、不確実性推定も提供することを目的としています。これは、意思決定者が様々なシナリオの下でリスクとトレードオフを評価するのに役立つ。確率的予測の手法には、ベイズ推定、生成的敵対ネットワーク（GAN）、変分オートエンコーダ（VAE）、正規化フロー[^5^][2]などがある。
- **多変量予測とマルチステップ予測**。多変量予測は、複数の変数や特徴を同時に予測することを扱い、マルチステップ予測は、複数の時間ステップを先に予測することを扱います。これらの課題には、高次元性、長期依存性、変数間の複雑な相互作用などの課題があります。多変量予測とマルチステップ予測には、ベクトル自己回帰モデル（VAR）、多出力回帰モデル（MOR）、配列間モデル（Seq2Seq）、エンコーダー・デコーダーモデル[^5^][2]があります。
- **メタ・ラーニング**。メタ学習は、データからどのように学習するかを学ぶことを目的とした学習パラダイムである。事前知識や経験を活用することで、異なるタスクやドメイン間での汎化性能を向上させることができる。メタ学習の手法には、少数点学習、転移学習、領域適応、MAML[^5^][2]などのメタ学習アルゴリズムがある。

# 将来の展望

時系列予測の今後の展望としては、以下のようなものがある。

- **プロンプトベースの学習**:プロンプトベースの学習は、GPT-3などの事前に学習した言語基盤モデルを活用して、時系列予測タスクを解決する新しいパラダイムである。数値の入力と出力は、自然言語モデルで処理可能なプロンプトに変換される。このアプローチは、ドメイン固有のデータの前処理や特徴量エンジニアリングの必要性を潜在的に減らすことができる[^6^][7]。
- **ハイブリッドモデル**:ハイブリッドモデルは、時系列予測の精度や頑健性を高めるために、異なる種類の手法やデータソースを組み合わせています。例えば、統計的手法と機械学習手法を統合したり、テキストや画像などの外部情報を数値データとともに利用したりすることができる[^5^][2]。
- 解釈可能なモデル**。解釈可能なモデルとは、モデルがどのように予測を行うか、あるいはなぜ誤りを犯すかについて、説明や洞察を提供することを目的としています。これは、ユーザーが予測を理解し、信頼し、改善するのを助けることができる。解釈可能なモデルは、注意メカニズム、顕著性マップ、反実仮想分析、因果推論[^5^][2]などの技法を用いることができる。

# プロンプトベース学習と時系列解析

- プロンプトベース学習は、事前に学習させた言語モデルにテキストによるプロンプトを与えて様々なタスクを実行させる自然言語処理のパラダイムである。プロンプトベース学習は、数発学習やゼロ発学習、知識抽出、テキスト生成などに用いることができる。
- 時系列解析とは、株価や天気、センサーなどの時間と共に変化するデータを解析することである。時系列分析は、予測、異常検知、分類などに利用することができる。
- プロンプトベース学習と時系列解析を組み合わせると、自然言語モデリングに基づく時系列データの表現学習を研究することができる。例えば、数値の入出力をプロンプトに変換する[^1^][1]、ラベルのないデータを用いてプロンプトの性能を向上させる[^2^][2]、あるいは時系列データの自然言語要約を生成するためにプロンプトベースの学習と時系列解析を利用することが可能である。

## モデル

- プロンプトベース学習と時系列解析を組み合わせることで、高い予測精度を実現しているモデルがある。
  - PromptCast。PromptCast：GPT-3やBARTなどの事前に学習した言語モデルを用いて、数値の入出力をプロンプトに変換し、時系列予測を行うモデルである。PromptCastは時系列データにおける長期的な依存関係や複雑なパターンを捉えることができます。PromptCastはCOVID-19の事例[^1^][1]、電力需要[^1^][1]などの予測に利用されている。
  - Co-training（コ・トレーニング）。Co-trainingとは、GPT-3やBARTなどの事前に学習させた言語モデルを用いて、ラベルのないデータを用いてプロンプトベースの学習を行うモデルである。Co-trainingは多様なプロンプトを生成し、情報量の多い例を選択することにより、プロンプトの性能を向上させることができる。Co-trainingは自然言語推論[^2^][2]、センチメント分析[^2^][2]などに利用されている。

# 熱力学と時系列解析

- 熱力学とは、物理系における熱、仕事、エネルギー、エントロピーの関係を研究する学問である。熱力学は、平衡、効率、可逆性など、さまざまな側面の測定や分析に利用できる。
- 時系列解析は、株価、天候パターン、センサー読み取り値など、時間とともに変化するデータの研究である。時系列解析は、予測、異常検出、分類などに利用することができる。
- 熱力学と時系列解析を組み合わせることで、熱力学の類推に基づく時系列データの統計的性質と挙動を研究することができる。例えば、エントロピー生産変動定理[^1^][1]などの統計法則を時系列データから導き出したり、エントロピー最大化[^2^][2]に基づいて時系列データの外れ値やトレンド変化を特定したり、熱平衡式[^3^][3]に基づいて原子炉の温度ダイナミクスをモデル化して推定するために熱力学や時系列解析を使用することができる。

## モデル

- 熱力学と時系列解析の組み合わせにより高い予測精度を持つモデルとして、以下のようなものがある。
  - エントロピー生産ゆらぎの定理(EPFT)。EPFTは、熱力学の原理を用いて、あらゆる定常過程におけるエントロピー生成率とその揺らぎの普遍的な関係を導き出すモデルである。EPFTは、時系列データの不可逆性とランダム性を捉えることができる。EPFTは、気候変動[^4^][4]、生化学反応[^5^][5]などの解析に用いられている。
  - エントロピーの最大化法（EMM）。EMMは、熱力学の原理を用いて、ある制約の下でエントロピーを最大化することにより、時系列をトレンド・季節性・ノイズ成分に分解するモデルである。EMMは非線形の複雑さを捉えることができる

# 数理最適化と時系列解析

- 数学的最適化とは、ある問題に対して、ある制約条件や基準に従って、考えられる選択肢の集合の中から最適な解を見つけ出す学問です。数理最適化は、効率性、実現可能性、トレードオフなど、さまざまな側面の測定や分析に使用することができます。
- 時系列解析は、株価、天候パターン、センサー読み取り値など、時間とともに変化するデータの研究である。時系列分析は、予測、異常検出、分類などに利用することができる。
- 数理最適化と時系列解析は、時系列モデルの最適なパラメータやコンポーネントを研究するために組み合わせることができます。例えば、時系列をトレンドとノイズの成分に分解したり[^1^][1]、自己回帰モデルの最適なラグ次数や係数を選択したり[^2^][2]、入れ子進化に基づいて非線形予測モデルを生成したり[^3^][3]など、数理最適化と時系列解析を組み合わせて使用することが可能です。

# 強化学習と時系列解析

- 強化学習（RL）は、環境との相互作用から学習し、報酬信号を最大化するための行動をとることを扱う機械学習の一分野である[^1^][4]。
- 時系列分析とは、株価、天候パターン、センサーの読み取り値など、時間と共に変化するデータを研究することである。時系列解析は、予測、異常検知、分類などに利用できる。
- RLは、予測を行動として扱い、状態の変化を現在の状態のみ（＋ランダム性）に依存させ、報酬を状態と行動に基づくものとすることで、時系列分析に適用することができる[^2^][1]。
- RLの時系列解析への応用例として、以下のようなものがある。
  - 金融時系列予測やポートフォリオ最適化。RLエージェントは、過去のデータと市場シグナルに基づいて、株式や暗号通貨の取引を学習することができる[^3^][3] [^4^][5]．
  - 時系列データ圧縮。RLエージェントは、高次元の時系列データを、重要な特徴を保ったまま低次元表現に圧縮することを学習することができる[^5^][6]。
  - 時系列異常の検出。RLエージェントは、能動学習とディープニューラルネットワーク[^6^][2]を用いて、実世界の時系列データにおける異常の検出を学習することが可能である。

## モデル

- RLと時系列解析を組み合わせて高い予測精度を持つモデルには、以下のようなものがある。
  - Deep Q-Networks (DQN)である。DQNは、RLにおける最適な行動-価値関数をディープニューラルネットワークで近似したモデルである。DQNは、画像や時系列データなどの高次元の入力を扱うことができる。DQNは金融時系列予測に利用されている[^3^][3]。
  - RNN（Recurrent Neural Networks）。RNNは、テキストや時系列データなどのシーケンシャルなデータを処理することができるモデルである。RNNは、データ中の長期的な依存関係を捉えることができる記憶メカニズムを持っています。RNNは、時系列予測に利用されている[^6^][2]。
  - Long Short-Term Memory (LSTM)。LSTMはRNNの一種で、RNNにおける勾配の消失や爆発という問題を克服できる、より複雑な記憶メカニズムを持っている。LSTMは、シーケンシャルなデータから長期的なパターンを学習することができる。LSTM は時系列予測や圧縮に利用されている[^6^][2] [^5^][6] 。

# 時系列予測

- 時系列予測とは、過去の観測結果に基づき、時間に依存する変数の将来値を予測するプロセスです。金融、経済、気候科学など様々な産業で広く利用されている。
- 時系列予測は、統計モデル、機械学習モデル、深層学習モデルなど、様々な方法を用いて行うことができる。それぞれの手法は、データの特性や予測の目的に応じて、長所と短所があります。
- 時系列予測に関する最近の研究・開発には、以下のようなものがある。

## 統計モデル

- 統計モデルは、データのパターンと傾向を記述する数学的方程式に基づいています。これらはしばしば、データの分布、定常性、季節性などについての仮定に依存しています。時系列予測によく使われる統計モデルには、ARIMA (AutoRegressive Integrated Moving Average), ETS (Exponential Smoothing), SARIMA (Seasonal ARIMA) 等があります。
- 統計モデルは実装も解釈も簡単ですが、複雑な非線形関係やデータの動的変化を捉えることができない場合があります。また、パラメータのチューニングやモデルオーダーの選択を手作業で行う必要がある場合もある。
- 最近の研究論文[^1^][1]では、STS (State Transition System) と呼ばれる時系列予測のための新しい統計モデルを提案している。STSは、システムの時間的な状態遷移を表現する確率的なグラフィカルモデルに基づいている。STSは複数の入力変数、非定常データ、欠損値、外れ値などを扱うことができる。STSはいくつかの実世界のデータセットにおいて、ARIMAやETSを凌駕する性能を発揮する。

## 機械学習モデル

- 機械学習モデルは、明示的なルールや方程式を使わずにデータから学習するアルゴリズムに基づくものである。機械学習モデルは、予測に必要な情報を取得するために、データの特徴や変換を使用することが多い。時系列予測のための一般的な機械学習モデルには、回帰木、ランダム・フォレスト（random forests)


# 時系列予測モデルの動向（2023年現在

時系列予測とは、ある変数の過去の観測値をもとに、将来の値を予測する作業である。金融、経済、エネルギー、健康、気候など様々な領域で多くの応用がある。近年、時系列予測は、大規模データの利用可能性と新しい手法やツールの開発により、大きな進歩を遂げている。

## 研究

時系列予測における現在の研究テーマには、以下のようなものがあります。

- **確率的予測**。ポイント予測だけでなく、不確実性区間や将来値の確率分布の推定を行うもの。確率論的予測は、意思決定やリスクマネジメントにおいて、より信頼性が高く、有益な予測を提供することができます。確率的予測手法の例としては、ベイズ推定[^1^][1]、分位点回帰[^2^][2]、深層生成モデル[^3^][3]がある。
- **多変量解析と階層的予測**。これは、複数の関連する変数を予測したり、異なる粒度の予測を集計したりするものです。多変量・階層予測は、変数間の複雑な依存関係や相互作用を捕らえ、正確さと一貫性を向上させることができます。多変量・階層予測手法の例としては、ベクトル自己回帰[^4^][4]、行列因子法[^5^][5]、グラフ・ニューラル・ネットワーク[^6^][6]があります。
- 説明可能な予測**。これは、予測に対して解釈可能な根拠や因果関係のメカニズムを提供することを含む。説明可能な予測はユーザーやステークホルダーの信頼と透明性を高めることができる。説明可能な予測手法の例としては、アテンション・メカニズム[^7^][7]、反実仮想分析[^8^][8]、因果推論[^9^][9]等がある。

## Forecasting Methods

実際に広く使われている、あるいは新たに登場した一般的な予測手法には、以下のようなものがあります。

- **統計モデル**。データのパターンや傾向をモデル化するために、数式や統計的仮定に依存する伝統的な方法である。統計モデルの例としては、自己回帰積分移動平均（ARIMA）、指数平滑化（ETS）、状態空間モデル（SSM）などがある。
- **機械学習モデル**。データ駆動型アルゴリズムや人工ニューラルネットワークを用いて、データから複雑な非線形関係を学習する最新の手法である。機械学習モデルの例として、サポートベクトル回帰（SVR）、ランダムフォレスト（RF）、長期短期記憶（LSTM）などがある。
- **ハイブリッドモデル**。統計モデルと機械学習モデルを組み合わせて、それぞれの強みを生かす手法である。ハイブリッドモデルの例として、ARIMA-LSTM、ETS-RF、SSM-LSTMなどがある[^18]。

# グラフニューラルネットワークと時系列解析

- グラフニューラルネットワーク（GNN）は、ソーシャルネットワーク、ナレッジグラフ、分子構造などのグラフ構造を持つデータに対して動作可能なニューラルネットワークの一種である。GNNは、グラフのノードの特徴とエッジの関係の両方から学習することができる。
- 時系列解析は、株価、天候パターン、センサーの読み取り値など、時間とともに変化するデータを研究するものである。時系列解析は、予測、異常検知、分類などに利用できる。
- GNNは、時系列データをグラフのノードとして扱い、空間情報や時間情報を用いてデータ間のエッジを構築することで、時系列分析に適用することができる。GNNは、多変量時系列データにおける複雑な依存関係やパターンを捉えることができる。

## モデル

- GNNと時系列解析を組み合わせて高い予測精度を持つモデルには、以下のようなものがある。
  - Graph Convolutional Networks (GCN)。GCN はグラフの畳み込みフィルタを用いて、隣接するノードの情報を集約するモデルです。GCN は、異なるタイプのノードとエッジを持つ異種グラフを扱うことができます。GCNは多変量時系列回帰[^1^][1]に利用されています。
  - グラフアテンションネットワーク (GAT)。GATはグラフ上の注目メカニズムを利用し、隣接するノードにその関連性に基づいて異なる重みを割り当てるモデルである。GATは、ノードの度数やエッジの重みが変化する動的なグラフを扱うことができる。GATは多変量時系列予測[^2^][2][^3^][3]に利用されている。
  - Temporal Graph Networks (TGN)。TGNは、グラフ上でリカレントニューラルネットワークを用い、ノードの特徴やエッジの関係の時間発展を捉えるモデルである。TGNはノードの状態やエッジの発生が変化する時間グラフを扱うことができる。TGNは動的リンク予測に利用されている[^4^][4]。

# 因果推論と時系列解析

- 因果推論とは，介入や治療がシステム内の結果にどのような影響を与えるかを研究することである．因果推論は、「ある政策の変更が経済に与える影響は何か」、「ある薬が患者の生存率に与える影響は何か」といった質問に答えるために使用することができる。
- 時系列分析とは、株価、天候パターン、センサーの読み取り値など、時間とともに変化するデータを研究することである。時系列解析は、予測、異常検知、分類などに利用できる。
- 因果推論と時系列分析を組み合わせることで、介入や治療の因果効果を時系列で研究することができる。例えば、因果推論と時系列分析を用いて、ロックダウン対策がCOVID-19の症例や死亡に与える影響を推定することができる。

グラフニューラルネットワーク（GNN）は、時系列予測モデルを作成するために使用できる人工ニューラルネットワークの一種です。GNNは、時系列データには時間的・空間的相関があるという考えに基づいている。GNNは、より正確な予測モデルを生成するために、これらの相関関係を識別し、利用することを学習することができる。GNNは短期と長期の両方の予測タスクに使用することができる。GNNはまた、複数の時系列を同時に予測するために使用することもできる。GNNはエネルギー需要予測、金融予測、医療診断など様々な用途に利用されている。

グラフニューラルネットワーク（GNN）は、社会ネットワーク、生物ネットワーク、交通ネットワークなどの複雑なシステムをモデル化する機械学習アプリケーションに広く利用されている。一方、時系列解析は、時間経過に伴うデータの予測・分析を扱う。GNNと時系列解析の組み合わせは、新しいモデルやライブラリの開発につながる活発な研究分野となっている。ここでは、この分野の最近の研究と関連リソースを箇条書きで紹介します。

研究しています。

T-GCN: ATemporalGraph Convolutional Network for Traffic Prediction (Liet al., 2018) - この論文では，GNNと時系列分析を組み合わせることで，交通予測のための時間グラフ畳み込みネットワーク（T-GCN）を提案している．提案モデルは、トラフィック予測タスクにおいて最先端のパフォーマンスを達成する。

STGCN:Spatio-TemporalGraph Convolutional Networks for TrafficForecasting(Yuet al., 2018) - この論文は、交通予測のための時空間グラフ畳み込みネットワーク（STGCN）を紹介するものである。このモデルは、トラフィックデータの空間的 依存性と時間的依存性の両方を捉えるように設計されている。

GTSです。A GraphTemporal SegmentationNetwork for Traffic Prediction (Maet al., 2021) - この論文は、交通予測のためのグラフ時間 セグメンテーションネットワーク（GTS）を提案するものである。このモデルは、グラフ畳み込みネットワークと時間的 セグメンテーションを統合し、交通データにおける空間的 依存性と時間的 依存性の両方を捕捉する。

図書館

PyTorchGeometric- PyTorch の幾何学的深層学習拡張ライブラリで、GNN とグラフ関連の演算をサポートする。ライブラリには様々な GNN モデルやデータセット、データの前処理や可視化のためのツールなどが含まれています。

DGL (Deep Graph Library) - これもグラフ関連操作のための深層学習ライブラリで、GNNや他のグラフベースのモデルをサポートしています。このライブラリは、様々なGNNモデルをサポートし、グラフの前処理や可視化のためのツールも提供します。

SOTAモデル。

MTGNN。多変量時系列予測のための多時間グラフニューラルネットワーク (Wu et al., 2020) - MTGNNは、多変量時系列予測を扱うことができる多時間グラフニューラルネットワークである。このモデルは階層構造を用いて、データ中の時間的依存性と系列横断的依存性の両方を捉える。
関連するリポジトリ

PyTorchForecasting- これは PyTorch ベースの時系列予測ライブラリで、GNN ベースのモデルを含む様々なモデルのサポートを提供しています。このライブラリには、データの前処理、モデルの学習、評価のためのツールが含まれています。

Time SeriesTransformer- 時系列解析のための様々な変換器ベースのモデルの PyTorch 実装を提供するリポジトリで、最近提案された変換器ベースの GNN モデルも含みます。

全体として、GNNと時系列解析の組み合わせは、複雑な機械学習の問題を解決する上で大きな可能性を示している。この分野の研究は現在も進行中であり、新しいモデルやライブラリが定期的に開発されている。GNNと時系列解析の利用は、交通予測、ソーシャルネットワーク解析、金融予測などの分野で幅広く応用されることが期待される。

時間グラフニューラルネットワークとは？

時空間グラフニューラルネットワーク（TGNN）は、時空間グラフニューラルネットワーク（STGNN）の一種で、時間的なイベントの列として表現される動的グラフに対して学習するために、時間の要素を考慮したものである。 TGNNは、深層学習モデルを非ユークリッド空間に拡張し、推薦システムやソーシャルネットワークなど様々なアプリケーションで最先端の性能を達成した強力なアルゴリズムであるGNN（Graph Neural Networks）の拡張版である。 しかし、GNNは静的なグラフ構造を仮定しているため、データが時間的に変化する場合には性能に限界がある。 この限界を克服するためにTGNNが提案され、いくつかの時間依存のアプリケーションにおいて、他の深層学習アルゴリズムと比較して優れた性能を達成している。 TGNNは、メモリモジュールとグラフベースの演算子の新しい組み合わせにより、動的なグラフ上で学習を行い、計算効率を高めながら、従来のアプローチを大幅に上回る性能を実現することができる。 TGNNは、時間的リンク予測、時間的グラフ生成、動的なノード分類などのタスクに使用されている。 TGNNは、PyTorch GeometricやTensorFlowなど様々な深層学習フレームワークを用いて実装されており、WikipediaやRedditなど様々なデータセットで評価されています。

時間グラフニューラルネットワーク
時間グラフニューラルネットワーク（TGNN）は、時間データの表現と処理にグラフを使用する新しいタイプのニューラルネットワーク技術である。このネットワークは、時間の経過に伴うオブジェクトやイベントの動的な関係を分析するためによく使用されます。TGNNは「ノード」の層で構成され、各ノードは特定の時間点またはイベントを表します。ノードはエッジで接続されており、エッジは時間経過に伴うノード間の関係を表す。TGNNの各層は独自のアルゴリズムあるいは「ルール」を持っており、ノードがどのように相互作用するかを決定する。ルールは通常、データから導き出され、ネットワークがイベント間の時間的関係を正確にモデル化できることを保証する。
TGNNは、株価の予測、顧客行動の分析、ネットワークトラフィックの異常の検出など、さまざまな問題に適用することができる。TGNNは、データをグラフで表現することで、イベント間の複雑かつ非線形な関係を捉えることができ、従来の機械学習手法よりも精度の高い結果を得ることができる。さらに、TGNNは大量のデータを迅速かつ効率的に処理するための学習が可能であり、リアルタイムアプリケーションに適している。
