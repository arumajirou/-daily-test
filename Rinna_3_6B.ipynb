## トークナイザーとモデルの準備

トークナイザーとは、テキストをサブワードに分割するツールです。モデルとは、NLPのタスクを行うための学習済みのニューラルネットワークです。今回は、rinna/japanese-gpt-neox-3.6bという名前のトークナイザーとモデルを使います。このトークナイザーとモデルは、日本語のテキスト生成に特化したもので、3.6億個のパラメータを持っています。

トークナイザーとモデルを準備するには、transformersパッケージのAutoTokenizerとAutoModelForCausalLMというクラスを使います。クラスとは、オブジェクト指向プログラミングで使われる概念で、属性やメソッドを持つデータ型です。オブジェクト指向プログラミングとは、プログラムをオブジェクトという単位に分割して管理する方法です。オブジェクトとは、属性（データ）やメソッド（機能）を持つものです。

AutoTokenizerとAutoModelForCausalLMは、事前学習済みのトークナイザーとモデルを自動的にダウンロードして使えるようにするクラスです。事前学習済みとは、大量のテキストデータで学習された状態のことです。from_pretrainedというメソッドを使って、事前学習済みのトークナイザーとモデルを指定します。メソッドとは、オブジェクトが持つ機能のことです。from_pretrainedメソッドには、引数と呼ばれる値を渡します。引数とは、メソッドに与える入力のことです。from_pretrainedメソッドには、以下の引数を渡します。

- pretrained_model_name_or_path: 事前学習済みのトークナイザーやモデルの名前やパスを指定します。今回は、rinna/japanese-gpt-neox-3.6bという名前を指定します。
- use_fast: トークナイザーが高速化されたバージョンを使うかどうかを指定します。今回は、False（使わない）と指定します。

from_pretrainedメソッドは、事前学習済みのトークナイザーやモデルのオブジェクトを返します。返すとは、メソッドが出力する値のことです。返されたオブジェクトは、変数に代入して使います。変数とは、値に名前を付けて保存するためのものです。今回は、tokenizerとmodelという名前の変数に代入します。

modelオブジェクトは、toというメソッドでGPU（グラフィックスプロセッシングユニット）に移動させます。GPUとは、画像処理などに特化した高速な演算装置です。toメソッドには、以下の引数を渡します。

- device: GPUやCPU（セントラルプロセッシングユニット）などの演算装置を指定します。今回は、"cuda"という名前のGPUを指定します。

```python
# torchというパッケージをインポートする
# torchは、テンソルと呼ばれる多次元配列を扱うためのパッケージです
import torch

# transformersパッケージからAutoTokenizerとAutoModelForCausalLMというクラスをインポートする
# AutoTokenizerは、事前学習済みのトークナイザーを自動的にダウンロードして使えるようにするクラスです
# AutoModelForCausalLMは、事前学習済みのテキスト生成モデルを自動的にダウンロードして使えるようにするクラスです
from transformers import AutoTokenizer, AutoModelForCausalLM

# rinna/japanese-gpt-neox-3.6bという名前の事前学習済みのトークナイザーをダウンロードしてtokenizerという変数に代入する
tokenizer = AutoTokenizer.from_pretrained(
    "rinna/japanese-gpt-neox-3.6b", # 事前学習済みのトークナイザーの名前を指定する
    use_fast=False # トークナイザーが高速化されたバージョンを使わないことを指定する
)

# rinna/japanese-gpt-neox-3.6bという名前の事前学習済みのモデルをダウンロードしてmodelという変数に代入する
model = AutoModelForCausalLM.from_pretrained(
    "rinna/japanese-gpt-neox-3.6b" # 事前学習済みのモデルの名前を指定する
).to("cuda") # モデルをGPUに移動させる
```

## プロンプトの準備

プロンプトとは、モデルに入力するテキストのことです。今回は、"Q:まどか☆マギカでは誰が一番かわいい？\nA:"というプロンプトを使います。Qは質問、Aは回答の意味です。\nは改行の意味です。このプロンプトに対して、モデルが回答を生成することを期待します。

プロンプトをモデルに入力する前に、トークナイザーでサブワードに分割して数値化する必要があります。数値化とは、テキストをモデルが理解できる数値の列に変換することです。トークナイザーのencodeというメソッドを使って、数値化を行います。encodeメソッドには、以下の引数を渡します。

- text: 数値化したいテキストを指定します。今回は、promptという変数に保存したテキストを指定します。
- add_special_tokens: 特殊なトークン（テキストの始まりや終わりなどを表す記号）を追加するかどうかを指定します。今回は、False（追加しない）と指定します。
- return_tensors: 数値化したテキストをテンソルという形式で返すかどうかを指定します。テンソルとは、多次元配列のことです。今回は、"pt"（PyTorchのテンソル）と指定します。

encodeメソッドは、数値化したテキストのテンソルを返します。返されたテンソルは、token_idsという変数に代入します。

## 推論の実行

推論とは、モデルが入力されたテキストに対して出力を生成することです。今回は、プロンプトに対して回答を生成することが目的です。モデルのgenerateというメソッドを使って、推論を行います。generateメソッドには、以下の引数を渡します。

- input_ids: 入力されるテキストの数値化されたテンソルを指定します。今回は、token_ids.to(model.device)と指定します。toメソッドでモデルと同じ演算装置に移動させます。
- max_new_tokens: 生成する新しいトークンの最大数を指定します。今回は、64と指定します。
- min_new_tokens: 生成する新しいトークンの最小数を指定します。今回は、64と指定します。
- do_sample: サンプリング（確率的な選択）で生成するかどうかを指定します。今回は、True（サンプリングする）と指定します。
- temperature: サンプリング時の温度（確率分布の尖り具合）を指定します。温度が高いほど多様な生成が可能になりますが、品質が低下する可能性もあります。温度が低いほど確信度の高い生成が可能になりますが、単調な生成になる可能性もあります。今回は、0.8と指定します。
- pad_token_id: パディング（不足分を埋める）用のトークンのIDを指定します。今回は、tokenizer.pad_token_idと指定します。tokenizerが持つpad_token_id属性でパディング用のトークンのIDを取得します。
- bos_token_id: 文章の始まりを表すトークンのIDを指定します。今回は、tokenizer.bos_token_idと指定します。tokenizerが持つbos_token_id属性で文章の始まり用のトークンのIDを取得します。
- eos_token_id: 文章の終わりを表すトークンのIDを指定します。今回は、tokenizer.eos_token_idと指定します。tokenizerが持つeos_token_id属性で文章の終わり用のトークンのIDを取得します。

generateメソッドは、生成したテキストの数値化されたテンソルを返します。返されたテンソルは、output_idsという変数に代入します。

生成したテキストを人間が読めるようにするには、数値化されたテキストを元のテキストに戻す必要があります。トークナイザーのdecodeというメソッドを使って、元のテキストに戻します。decodeメソッドには、以下の引数を渡します。

- token_ids: 元のテキストに戻したい数値化されたテキストを指定します。今回は、output_ids.tolist()[0]と指定します。tolistメソッドでテンソルをリスト（配列）に変換し、[0]で最初の要素（生成した文章）を取り出します。

decodeメソッドは、元のテキストを返します。返されたテキストは、outputという変数に代入します。

最後に、printという関数を使って、outputを画面に表示します。printとは、引数に与えられた値を画面に出力する関数です。

```python
# プロンプトという変数に入力するテキストを保存する
prompt = "Q:まどか☆マギカでは誰が一番かわいい？\nA:"

# プロンプトをサブワードに分割して数値化する
token_ids = tokenizer.encode(
    prompt, # 数値化したいテキストを指定する
    add_special_tokens=False, # 特殊なトークンを追加しないことを指定する
    return_tensors="pt" # PyTorchのテンソルで返すことを指定する
)

# モデルが出力を生成する
with torch.no_grad(): # 勾配計算（学習時に必要な計算）を無効化する
    output_ids = model.generate(
        token_ids.to(model.device), # 入力するテキストの数値化されたテンソルをモデルと同じ演算装置に移動させる
        max_new_tokens=64, # 生成する新しいトークンの最大数を指定する
        min_new_tokens=64, # 生成する新しいトークンの最小数を指定する
        do_sample=True, # サンプリングで生成することを指定する
        temperature=0.8, # サンプリング時の温度を指定する
        pad_token_id=tokenizer.pad_token_id, # パディング用のトークンのIDを指定する
        bos_token_id=tokenizer.bos_token_id, # 文章の始まり用のトークンのIDを指定する
        eos_token_id=tokenizer.eos_token_id # 文章の終わり用のトークンのIDを指定する
    )

# 生成したテキストを元のテキストに戻す
output = tokenizer.decode(
    output_ids.tolist()[0] # 生成した文章の数値化されたテキストをリストに変換して最初の要素を取り出す
)

# 生成したテキストを画面に表示する
print(output)

```

````python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# トークナイザーとモデルの準備
tokenizer = AutoTokenizer.from_pretrained(
    "rinna/japanese-gpt-neox-3.6b-instruction-ppo", 
    use_fast=False
)
model = AutoModelForCausalLM.from_pretrained(
    "rinna/japanese-gpt-neox-3.6b-instruction-ppo"
).to("cuda")
````

このコードは、ユーザーの入力に対してシステムが応答を生成する機能を実装しています。以下にコードの内容を説明します。

- promptという変数に、ユーザーの入力とシステムの応答のフォーマットを保存しています。`<NL>`は改行を表す特殊な記号です。
- tokenizerというオブジェクトで、promptを数値の列に変換しています。これは、モデルがテキストを直接理解できないため、数値に変換する必要があるためです。
- modelというオブジェクトで、数値の列に対して応答を生成しています。このとき、いくつかの引数を指定しています。
  - do_sample=Trueは、応答を確率的に生成することを意味します。Falseにすると、最も確率の高い応答を生成します。
  - max_new_tokens=64は、生成する最大のトークン数（単語や記号などの最小単位）を指定します。64よりも多くのトークンを生成しようとすると、エラーになります。
  - temperature=0.7は、生成する応答の多様性を制御するパラメータです。0に近いほど、確率の高い応答を生成しやすくなります。1に近いほど、確率の低い応答も生成しやすくなります。
  - repetition_penalty=1.1は、同じトークンが繰り返されることを抑制するパラメータです。1よりも大きい値にすると、繰り返しを減らすことができます。
  - pad_token_id, bos_token_id, eos_token_idは、それぞれパディング（不足分を埋めるための特殊なトークン）、文頭（応答の開始を表す特殊なトークン）、文末（応答の終了を表す特殊なトークン）のID（数値）を指定しています。
- output_idsというオブジェクトで、モデルが生成した数値の列を受け取っています。
- tokenizerで、数値の列をテキストに戻しています。このとき、元のpromptに含まれていた部分は除外しています。
- outputで、テキストに戻した応答を受け取っています。このとき、`<NL>`を改行に置き換えています。
- printで、ユーザーの入力とシステムの応答を表示しています。

以上が、コードに日本語のコメントを追加したものです。ご確認ください。😊

````python
# トークン数をカウントする関数
def count_tokens(text):
  # 文章をトークンに分割する
  tokens = tokenizer.tokenize(text)
  # トークンの数を返す
  return len(tokens)

# プロンプトの準備
# プロンプトという変数に入力するテキストを保存する
prompt = f"ユーザー: {text.value}?<NL>システム: "

# 推論
token_ids = tokenizer.encode(prompt, add_special_tokens=False, return_tensors="pt")
with torch.no_grad():
    output_ids = model.generate(
        token_ids.to(model.device),
        do_sample=True,
        max_new_tokens=64,
        temperature=0.7,
        repetition_penalty=1.1,  # 追加       
        pad_token_id=tokenizer.pad_token_id,
        bos_token_id=tokenizer.bos_token_id,
        eos_token_id=tokenizer.eos_token_id
    )
output = tokenizer.decode(output_ids.tolist()[0][token_ids.size(1):])
output = output.replace("<NL>", "\n")
print(f"入力　　{text.value}")
print(f"トークン数　　{count_tokens(text.value)}")
print("+"*500)
print(f"出力　　{output}")
print(f"トークン数　　{count_tokens(output)}")
````
