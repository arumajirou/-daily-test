# ChatDocsとは？（What is ChatDocs?）

ChatDocsは、AI（人工知能）を使って、オフラインで文書と会話できるツールです。オフラインとは、インターネットにつながらなくても使えるという意味です。このツールを使うと、文書の中にある情報を簡単に見つけたり、まとめたり、質問したりできます。例えば、学校の宿題で調べ物をするときに、たくさんの文書を読む必要がなくなります。ChatDocsは、PrivateGPTというAIモデル（AIの仕組み）に基づいていますが、もっといろいろな機能があります。

## ChatDocsの使い方（How to use ChatDocs）

- まず、このツールとAIモデルをインターネットからダウンロードしてインストールします。それ以外は、インターネットにつながらなくても大丈夫です。
- 次に、調べたい文書をChatDocsに追加します。文書はdbというフォルダに保存されます。
- そして、ブラウザ（インターネットを見るためのソフト）でhttp://localhost:5000というアドレスにアクセスします。そこで、Web UI（ウェブ・ユーアイ）という画面が見えます。Web UIは、ChatDocsと会話するための画面です。
- Web UIで、文書に関する質問を入力します。すると、ChatDocsが答えてくれます。答えには、文書の中から引用した部分も含まれます。引用した部分をクリックすると、文書のどこにあるか確認できます。

## ChatDocsの設定方法（How to configure ChatDocs）

- ChatDocsの設定は、chatdocs.ymlというファイルで変更できます。このファイルは、自分で作って保存します。ファイルの中には、設定したい項目だけ書けばいいです。全部書く必要はありません。設定できる項目は、デフォルトのchatdocs.ymlファイルを参考にしてください。
- 設定を変更したら、ChatDocsを再起動してください。
- 設定できる項目の一部を紹介します。

### embeddings_model（エンベディングス・モデル）
- エンベディングス・モデルとは、文書の内容を数値に変換するAIモデルです。数値に変換することで、文書の類似度（似ている度合い）や重要度（大事な度合い）を計算できます。
- エンベディングス・モデルは、chatdocs.ymlファイルで変更できます。以下のように書きます。

```
embeddings_model:
  name: sentence_transformers/msmarco-distilbert-base-v4
```

- nameの後ろには、使いたいエンベディングス・モデルの名前を書きます。名前はここで探せます：https://huggingface.co/models?filter=pytorch
- エンベディングス・モデルを変更したら、dbというフォルダを消して、文書をもう一度追加してください。

### c_transformers_model（シー・トランスフォーマーズ・モデル）
- シー・トランスフォーマーズ・モデルとは、文書に関する質問に答えるAIモデルです。このモデルは、C Transformers GGMLという技術（技術の説明はここにあります：https://github.com/THUDM/CogQA）を使っています。
- シー・トランスフォーマーズ・モデルは、chatdocs.ymlファイルで変更できます。以下のように書きます。

```
c_transformers_model:
  name: ctransformers/distilbert-base-cased-distilled-squad
```

- nameの後ろには、使いたいシー・トランスフォーマーズ・モデルの名前を書きます。名前はここで探せます：https://huggingface.co/models?filter=pytorch
- シー・トランスフォーマーズ・モデルを変更したら、chatdocs downloadというコマンド（命令）を実行して、モデルをダウンロードしてください。

### transformers_model（トランスフォーマーズ・モデル）
- トランスフォーマーズ・モデルとは、文書に関する質問に答えるAIモデルです。このモデルは、Transformersという技術（技術の説明はここにあります：https://huggingface.co/transformers/）を使っています。
- トランスフォーマーズ・モデルを使うには、chatdocs.ymlファイルに以下のように書きます。

```
use_transformers: true
```

- use_transformersの後ろには、true（使う）かfalse（使わない）を書きます。
- トランスフォーマーズ・モデルは、chatdocs.ymlファイルで変更できます。以下のように書きます。

```

  name: deepset/roberta-base-squad2
```

- nameの後ろには、使いたいトランスフォーマーズ・モデルの名前を書きます。名前はここで探せます：https://huggingface.co/models?filter=pytorch
- トランスフォーマーズ・モデルを変更したら、chatdocs downloadというコマンド（命令）を実行して、モデルをダウンロードしてください。

### gptq_model（ジーピーティーキュー・モデル）
- ジーピーティーキュー・モデルとは、文書に関する質問に答えるAIモデルです。このモデルは、GPTQという技術（技術の説明はここにあります：https://github.com/marella/auto-gptq）を使っています。
- ジーピーティーキュー・モデルを使うには、chatdocs.ymlファイルに以下のように書きます。

```
use_gptq: true
```

- use_gptqの後ろには、true（使う）かfalse（使わない）を書きます。
- ジーピーティーキュー・モデルは、chatdocs.ymlファイルで変更できます。以下のように書きます。

```
