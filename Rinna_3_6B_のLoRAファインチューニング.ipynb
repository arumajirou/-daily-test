## Rinna-3.6Bとは何か？
- Rinna-3.6Bは、日本語の文章を生成したり理解したりすることができるプログラムです。（文章を生成するというのは、例えば「今日の天気は？」と聞かれたら「晴れです」と答えるようなことです。文章を理解するというのは、例えば「晴れです」という答えから「今日は外で遊べる」ということがわかるようなことです。）
- Rinna-3.6Bは、日本語の文章がたくさん入ったデータベース（データベースとは、情報を整理して保存する場所のことです）から学習しています。（学習するというのは、例えば「今日の天気は？」という質問に対して、「晴れです」という答えが正しいことを覚えるようなことです。）
- Rinna-3.6Bは、36億個ものパラメーター（パラメーターとは、プログラムの動き方や性能を決める数字や設定のことです）を持っています。パラメーターが多いほど、プログラムは複雑なことができますが、その分学習に時間やコンピューターの力が必要になります。

## LoRAとは何か？
- LoRAは、言語モデル（言語モデルとは、文章を生成したり理解したりするプログラムのことです）の性能を向上させる技術です。（性能を向上させるというのは、例えば「今日の天気は？」という質問に対して、「晴れです」という答えだけでなく、「気温は25度で湿度は50%です」というようにもっと詳しく答えられるようにするようなことです。）
- LoRAは、言語モデルのパラメーターを少し変えることで、特定の目的に合わせて言語モデルを調整します。（特定の目的に合わせて調整するというのは、例えば「今日の天気は？」という質問に対して、「晴れです」という答えではなく、「雨です」という答えが必要な場合に、パラメーターを変えて「雨です」と答えられるようにするようなことです。）
- LoRAは、パラメーターを少し変えるだけなので、学習に時間やコンピューターの力があまり必要ありません。

## Google Colabとは何か？
- Google Colabは、オンラインでコードを実行できるクラウドサービスです。（コードとは、プログラムを作るための言語のことです。クラウドサービスとは、インターネットを通して利用できるサービスのことです。）

````python
# transformersモジュールからAutoTokenizerというクラスをインポートする
# AutoTokenizerとは、モデルに合わせて自動的にトークナイザーを選択するためのクラスである
# トークナイザーとは、テキストをトークンと呼ばれる単位に分割するためのツールである
from transformers import AutoTokenizer

# トークナイザーの準備
# tokenizerという変数に、AutoTokenizerのインスタンスを代入する
# from_pretrainedというメソッドは、事前学習済みのトークナイザーをロードするためのものである
# model_nameという変数には、使用するモデルの名前が代入されている
# use_fast=Falseというオプションは、高速化されたトークナイザーを使用しないことを指定するためのものである
tokenizer: AutoTokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)
````

````python
import transformers # transformersというライブラリを使えるようにする
eval_steps = 200 # 評価を行うステップ数を200にする
save_steps = 200 # モデルを保存するステップ数を200にする
logging_steps = 20 # ログを出力するステップ数を20にする

# トレーナーの準備
trainer = transformers.Trainer( # Trainerというクラスのインスタンス（実体）をtrainerという名前で作る
    model=model, # 学習させるモデルをmodelという変数に入れる
    train_dataset=train_data, # 学習に使うデータセットをtrain_dataという変数に入れる
    eval_dataset=val_data, # 評価に使うデータセットをval_dataという変数に入れる
    args=transformers.TrainingArguments( # 学習の設定をTrainingArgumentsというクラスで作る
        num_train_epochs=3, # 学習のエポック数（全てのデータを一回学習すること）を3にする
        learning_rate=3e-4, # 学習率（パラメータの更新量）を0.0003にする
        logging_steps=logging_steps, # ログを出力するステップ数をlogging_stepsという変数に入れる
        evaluation_strategy="steps", # 評価のタイミングをステップ数で決める
        save_strategy="steps", # モデルの保存のタイミングをステップ数で決める
        eval_steps=eval_steps, # 評価を行うステップ数をeval_stepsという変数に入れる
        save_steps=save_steps, # モデルを保存するステップ数をsave_stepsという変数に入れる
        output_dir=output_dir, # モデルやログなどの出力先のディレクトリ（フォルダ）をoutput_dirという変数に入れる
        report_to="none", # レポート（学習の結果など）の送信先をなしにする
        save_total_limit=3, # 保存するモデルの最大数を3にする
        push_to_hub=False, # Hugging Face Hub（モデルやデータセットなどの共有サイト）にモデルを送信しない
        auto_find_batch_size=True # バッチサイズ（一度に学習するデータの個数）を自動的に見つける
    ),
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False), 
    # data_collatorとは、複数のデータからバッチ（一つのまとまり）を作るための関数です。
    # transformers.DataCollatorForLanguageModelingは、言語モデリング（文や単語などの生成）用のdata_collatorです。
    # tokenizerは、文や単語などを数値に変換するためのツールです。
    # mlmは、Masked Language Modeling（一部の単語を隠して予測する技術）を使うかどうかです。ここでは使わない。
)
````
