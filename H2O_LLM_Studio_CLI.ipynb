## H2O LLM Studio GUIの起動

H2O LLM Studio GUIは、以下の方法で起動できます¹。

- requirements.txtを使ってインストールする方法
- Dockerからナイトリービルドを使って起動する方法
- Dockerイメージを自分でビルドして起動する方法

### requirements.txtを使ってインストールする方法

requirements.txtは、H2O LLM Studioが動作するために必要なプログラムやライブラリ（部品）の一覧です。これを使って、簡単にインストールできます。

まず、H2O LLM Studioのソースコード（プログラムの元）をダウンロードします。以下のコマンドを実行してください。

```bash
git clone https://github.com/h2oai/h2o-llmstudio.git
```

次に、ダウンロードしたフォルダに移動して、requirements.txtを使ってインストールします。以下のコマンドを実行してください。

```bash
cd h2o-llmstudio
pip install -r requirements.txt
```

### Dockerからナイトリービルドを使って起動する方法

Dockerとは、コンピューター上に仮想的な環境（コンテナ）を作って、プログラムを動かすためのツールです。Dockerを使うと、インストールや設定などの手間が省けます。

ナイトリービルドとは、開発中の最新版のプログラムを毎日自動的にビルド（作成）することです。ナイトリービルドを使うと、最新の機能や修正が利用できますが、不安定になる可能性もあります。

まず、Dockerをインストールしてください。以下のURLからダウンロードできます。

https://docs.docker.com/get-docker/

次に、以下のコマンドを実行して、H2O LLM StudioのナイトリービルドのDockerイメージ（コンテナの元）をダウンロードします。

```bash
docker pull h2oai/llmstudio:nightly
```

次に、以下のコマンドを実行して、H2O LLM Studio GUIを起動します。

```bash
docker run --rm -p 10101:10101 h2oai/llmstudio:nightly
```

### Dockerイメージを自分でビルドして起動する方法

Dockerイメージを自分でビルドすると、カスタマイズやデバッグ（不具合の調査）ができますが、時間がかかる場合があります。

まず、H2O LLM Studioのソースコードをダウンロードします。以下のコマンドを実行してください。

```bash
git clone https://github.com/h2oai/h2o-llmstudio.git
```

次に、ダウンロードしたフォルダに移動して、Dockerイメージをビルドします。以下のコマンドを実行してください。

```bash
cd h2o-llmstudio
docker build -t h2oai/llmstudio:latest .
```

次に、以下のコマンドを実行して、H2O LLM Studio GUIを起動します。

```bash
docker run --rm -p 10101:10101 h2oai/llmstudio:latest
```

## データ形式

H2O LLM Studioでは、微調整に使うデータは、以下の形式で用意する必要があります 。

- テキストファイル（.txt）であること
- 一行に一つのテキストが入っていること
- テキストはタブ（\t）で区切られた二つの部分からなること
- 最初の部分は入力（質問やプロンプトなど）であること
- 次の部分は出力（答えや生成文など）であること

例えば、以下のようなデータがあります。

```
What is the capital of Japan?	Tokyo
Who wrote Harry Potter?	J.K. Rowling
How many planets are there in the solar system?	Eight
```

このデータは、LLMに質問に答えさせるための微調整に使えます。最初の部分が質問で、次の部分が答えです。

データを用意したら、H2O LLM Studio GUIのNew Experimentボタンをクリックして、実験を開始します。

### 実験の開始

New Experimentボタンをクリックすると、以下のような画面が表示されます 。

![New Experiment](https://docs.h2o.ai/h2o-llmstudio/_images/llm_studio_new_experiment.png)

この画面では、以下のことができます 。

- Experiment Nameで、実験に名前を付けることができます。
- Model Typeで、微調整するLLMの種類を選択できます。現在はGPT-NeoとGPT-Jが利用できます。
- Model Sizeで、微調整するLLMのサイズを選択できます。サイズが大きいほど性能は高くなりますが、学習に時間やメモリがかかります。
- Data Fileで、微調整に使うデータファイルをアップロードできます。
- Hyperparametersで、微調整に使うパラメーターを設定できます。パラメーターについては後述します。
- Startボタンで、実験を開始できます。

パラメーターには以下のものがあります 。

| パラメーター | 説明 |
| --- | --- |
| Batch Size | 一度に学習するデータの数 |
| Learning Rate | 学習率（学習の速さ） |
| Num Epochs | エポック数（全データを何回学習するか） |
| Max Length | 入力や出力の最大文字数 |
| Min Length | 出力の最小文字数 |
| Early Stopping | 早期終了（性能が改善しなくなったら学習を止めるか） |
| Patience | 早期終了の基準（何回連続で性能が改善しなかったら学習を止めるか） |
| Gradient Accumulation Steps | 勾配蓄積ステップ数（何回の学習で勾配を更新するか） |
| Weight Decay | 重み減衰（過学習を防ぐために重みを減らすか） |
| Adam Epsilon | アダムイプシロン（勾配の更新に使う微小な値） |
| Warmup Steps | ウォームアップステップ数（学習率を徐々に上げるステップ数） |
| Max Grad Norm | 最大勾配ノルム（勾配の大きさの上限） |
| Top K | トップK（出力の候補を上位K個に絞るか） |
| Top P | トップP（出力の候補を確率P以上に絞るか） |
| Repetition Penalty | 繰り返しペナルティ（出力に同じ単語が繰り返されることを減らすか） |
| Temperature | 温度（出力の多様性を増やすか） |
| No Repeat N-Gram Size | 繰り返し禁止Nグラムサイズ（出力に同じNグラムが繰り返されないようにするか） |
| Bad Words Ids | 禁止単語ID（出力に含めない単語のIDのリスト） |
| Num Beams | ビーム数（出力の候補を何本の枝で探索するか） |
| Length Penalty | 長さペナルティ（出力の長さに応じてスコアを変えるか） |
| Diversity Penalty | 多様性ペナルティ（ビーム内で出力の多様性を増やすか） |

パラメーターの詳しい説明や効果は、以下のURLを参照してください。

https://huggingface.co/transformers/main_classes/model.html#transformers.generation_utils.GenerationMixin.generate

パラメーターを設定したら、Startボタンをクリックして実験を開始します。

## OASSTデータでCLI経由で実行する例

H2O LLM Studioでは、GUIだけでなく、CLI（コマンドラインインターフェース）でもLLMを微調整できます。CLIとは、コマンド（文字列）を入力して操作する方法です。

CLIを使うと、GUIにない機能や設定が利用できます。例えば、Neptuneというサービスと連携して実験を管理したり、GPU（高速な計算装置）を使って学習を高速化したりすることができます。

ここでは、OASSTデータという日本語の質問応答データセットを使って、GPT-J 6Bモデルを微調整する例を紹介します。

### データの準備

まず、OASSTデータをダウンロードします。以下のURLからダウンロードできます。

https://github.com/nlp-lab-oit/oasst

次に、ダウンロードしたデータからtrain.txtファイルを抽出します。このファイルには、質問と答えのペアが入っています。

しかし、このファイルはH2O LLM Studioに使える形式ではありません。H2O LLM Studioに使える形式に変換するためには、以下のスクリプト（プログラム）を実行します。

```python
import json

with open("train.txt", "r", encoding="utf-8") as f:
    data = json.load(f)

with open("oasst.txt", "w", encoding="utf-8") as f:
    for item in data:
        question = item["question"]
        answer = item["answer"]
        f.write(f"{question}\t{answer}\n")
```

このスクリプトを実行すると、oasst.txtというファイルが作成されます。このファイルは、H2O LLM Studioに使える形式になっています。

### モデルのダウンロード

次に、微調整するLLMをダウンロードします。ここでは、GPT-J 6Bというモデルを使います。以下のコマンドを実行してください。

```bash
wget https://the-eye.eu/public/AI/GPT-J-6B/step_383500_slim.tar.zstd
```

このコマンドで、step_383500_slim.tar.zstdというファイルがダウンロードされます。このファイルは、圧縮されたモデルのファイルです。

圧縮を解除するためには、以下のコマンドを実行してください。

```bash
tar -I zstd -xvf step_383500_slim.tar.zstd
```

このコマンドで、step_383500_slimというフォルダが作成されます。このフォルダには、モデルのファイルが入っています。

### Neptuneの設定

Neptuneとは、実験の管理や可視化をするためのサービスです。Neptuneを使うと、実験の履歴や結果を保存したり、グラフや表で見たりすることができます。

Neptuneを使うには、以下の手順が必要です 。

- Neptuneアカウントを作成する
- Neptuneプロジェクトを作成する
- Neptune APIトークンを取得する
- Neptune APIトークンを環境変数に設定する

#### Neptuneアカウントを作成する

まず、Neptuneアカウントを作成します。以下のURLから作成できます。

https://neptune.ai/

#### Neptuneプロジェクトを作成する

次に、Neptuneプロジェクトを作成します。プロジェクトとは、実験の集まりのことです。以下のURLから作成できます。

https://app.neptune.ai/projects/create-project

#### Neptune APIトークンを取得する

次に、Neptune APIトークンを取得します。APIトークンとは、Neptuneと連携するために必要なパスワードのようなものです。以下のURLから取得できます。

https://app.neptune.ai/get_my_api_token

#### Neptune APIトークンを環境変数に設定する

次に、Neptune APIトークンを環境変数に設定します。環境変数とは、コマンドラインで使える変数のことです。以下のコマンドを実行してください。

```bash
export NEPTUNE_API_TOKEN=YOUR_API_TOKEN
```

ここで、YOUR_API_TOKENは取得したAPIトークンに置き換えてください。

### モデルの微調整

以上の準備ができたら、モデルの微調整を開始します。以下のコマンドを実行してください。

```bash
python train.py \
--model_type gptj \
--model_size 6B \
--model_path step_383500_slim \
--data_file oasst.txt \
--batch_size 1 \
--learning_rate 5e-5 \
--num_epochs 3 \
--max_length 128 \
--min_length 1 \
--early_stopping True \
--patience 3 \
--gradient_accumulation_steps 16 \
--weight_decay 0.01 \
--adam_epsilon 1 \
--warmup_steps 200 \
--max_grad_norm 1.0 \
--top_k 0 \
--top_p 0.9 \
--repetition_penalty 1.0 \
--temperature 1.0 \
--no_repeat_ngram_size 3 \
--bad_words_ids [] \
--num_beams 1 \
--length_penalty 1.0 \
--diversity_penalty 0.0 \
--neptune_project YOUR_PROJECT_NAME
```

ここで、YOUR_PROJECT_NAMEは作成したNeptuneプロジェクトの名前に置き換えてください。

このコマンドでは、以下のパラメーターを設定しています。

| パラメーター | 値 | 説明 |
| --- | --- | --- |
| model_type | gptj | モデルの種類はGPT-J |
| model_size | 6B | モデルのサイズは6B |
| model_path | step_383500_slim | モデルのファイルが入っているフォルダの名前 |
| data_file | oasst.txt | データファイルの名前 |
| batch_size | 1 | 一度に学習するデータの数は1 |
| learning_rate | 5e-5 | 学習率は5e-5 |
| num_epochs | 3 | エポック数は3 |
| max_length | 128 | 入力や出力の最大文字数は128 |
| min_length | 1 | 出力の最小文字数は1 |
| early_stopping | True | 早期終了を有効にする |
| patience | 3 | 早期終了の基準は3回 |
| gradient_accumulation_steps | 16 | 勾配蓄積ステップ数は16 |
| weight_decay | 0.01 | 重み減衰は0.01 |
| adam_epsilon | 1e-8 | アダムイプシロンは1e-8 |
| warmup_steps | 200 | ウォームアップステップ数は200 |
| max_grad_norm | 1.0 | 最大勾配ノルムは1.0 |
| top_k | 0 | トップKは無効にする |
| top_p | 0.9 | トップPは0.9 |
| repetition_penalty | 1.0 | 繰り返しペナルティは無効にする |
| temperature | 1.0 | 温度は無効にする |
| no_repeat_ngram_size | 3 | 繰り返し禁止Nグラムサイズは3 |
| bad_words_ids | [] | 禁止単語IDは無効にする |
| num_beams | 1 | ビーム数は1 |
| length_penalty | 1.0 | 長さペナルティは無効にする |
| diversity_penalty | 0.0 | 多様性ペナルティは無効にする |
| neptune_project | YOUR_PROJECT_NAME | Neptuneプロジェクトの名前 |

コマンドを実行すると、以下のような出力が表示されます。

## モデルチェックポイント

モデルチェックポイントとは、モデルの学習の途中経過を保存することです。モデルチェックポイントを使うと、以下のメリットがあります。

- 学習が途中で中断されても、最後に保存したチェックポイントから再開できます。
- 学習が終了した後でも、途中のチェックポイントを比較したり評価したりすることができます。
- 学習が過学習（性能が悪化する現象）してしまった場合でも、過学習しなかったチェックポイントを使うことができます。

H2O LLM Studioでは、以下の方法でモデルチェックポイントを使うことができます。

- GUIで実験を保存すると、チェックポイントが自動的に保存されます。
- CLIで実験を行うと、エポックごとにチェックポイントが自動的に保存されます。
- CLIで実験を再開するときに、--model_pathでチェックポイントのフォルダを指定すると、そのチェックポイントから再開できます。

````python
# このコードは、numpyとpandasという名前のライブラリをインポートするためのものです。
# numpyとは、数値計算や行列操作などを高速に行うためのライブラリです。
# pandasとは、データ分析や操作などを容易に行うためのライブラリです。
# インポートとは、他のファイルやモジュールなどを現在のプログラムで使えるようにすることです。
import numpy as np # numpyをnpという名前でインポートします。これにより、numpyの機能をnpという名前で呼び出せます。
import pandas as pd # pandasをpdという名前でインポートします。これにより、pandasの機能をpdという名前で呼び出せます。

# このコードは、pandasのオプションを設定するためのものです。
# オプションとは、プログラムの動作や表示などを変更するための設定値のことです。
pd.options.mode.chained_assignment = None # chained_assignmentというオプションをNoneに設定します。これにより、データフレームの一部を切り出して代入する際に警告メッセージが表示されなくなります。データフレームとは、表形式のデータを扱うための構造のことです。

# このコードは、datasetsという名前のライブラリをインポートし、load_datasetという名前の関数を使ってOpenAssistant/oasst1という名前のデータセットを読み込むためのものです。
# datasetsとは、機械学習や自然言語処理などに使えるデータセットを提供するライブラリです。
# 関数とは、特定の処理や計算などをまとめたもので、名前や引数などを指定して呼び出せるものです。
# データセットとは、機械学習や分析などに使うデータの集まりのことです。
from datasets import load_dataset # datasetsからload_datasetという関数をインポートします。これにより、load_datasetという関数を使えるようになります。

ds = load_dataset("OpenAssistant/oasst1") # load_dataset関数に"OpenAssistant/oasst1"という文字列（引数）を渡して呼び出し、返り値（戻り値）をdsという変数に代入します。文字列とは、""や''で囲まれた文字や記号のことです。引数とは、関数に渡す値や情報のことです。返り値（戻り値）とは、関数が呼び出された後に返す値や情報のことです。変数とは、値や情報を一時的に保存しておくための箱（メモリ領域）のことです。

train = ds['train'] # ds変数（データセット）から'train'というキー（名前）に対応する値（データ）を取り出し、trainという変数に代入します。キーとは、辞書型と呼ばれるデータ構造の一つで、キーと値のペアでデータを管理するものです。辞書型とは、{}で囲まれたデータ構造のことです。
val = ds['validation'] # ds変数（データセット）から'validation'というキー（名前）に対応する値（データ）を取り出し、valという変数に代入します。

train = pd.DataFrame(train) # train変数（データ）をpd.DataFrame関数に渡して呼び出し、返り値（データフレーム）をtrain変数に再代入します。これにより、train変数のデータ構造がデータフレームに変換されます。
val = pd.DataFrame(val) # val変数（データ）をpd.DataFrame関数に渡して呼び出し、返り値（データフレーム）をval変数に再代入します。これにより、val変数のデータ構造がデータフレームに変換されます。

# このコードは、prep_dataという名前の関数を定義するためのものです。
# prep_data関数は、引数としてdfという名前のデータフレームを受け取り、必要な処理を行って返り値としてdf_assistantという名前のデータフレームを返す関数です。
# 関数の定義とは、関数の名前や引数や処理や返り値などを指定することです。
def prep_data(df): # defというキーワードを使って、prep_dataという名前の関数を定義します。引数としてdfという名前のデータフレームを受け取ることを指定します。キーワードとは、Pythonで特別な意味や役割を持つ単語のことです。
    df_assistant = df[(df.role=="assistant") & (df["rank"]==0.0)] # df変数（データフレーム）からrole列が"assistant"でrank列が0.0である行だけを抽出し、df_assistantという変数に代入します。列とは、データフレームの縦方向の部分のことです。行とは、データフレームの横方向の部分のことです。抽出とは、条件に合う部分だけを取り出すことです。
    df_prompter = df[(df.role=="prompter")] # df変数（データフレーム）からrole列が"prompter"である行だけを抽出し、df_prompterという変数に代入します。
    df_prompter = df_prompter.set_index("message_id") # df_prompter変数（データフレーム）のインデックス（行番号）をmessage_id列に設定し、df_prompter変数に再代入します。インデックスとは、データフレームの各行に付けられた番号やラベルのことです。
    df_assistant["output"] = df_assistant["text"].values # df_assistant変数（データフレーム）にoutput列を追加し、text列の値を代入します。追加とは、新しい要素や項目などを加えることです。values属性とは、データフレームやシリーズなどから値だけを取り出すためのものです。シリーズとは、一次元配列のようなデータ構造で、一つの列や行などを表すことができます。

    inputs = [] # inputsという名前の空のリストを作成し、inputsという変数に代入します。リストとは、[]で囲まれたデータ構造で、複数の要素を順番に並べたものです。空とは、要素がないことです。
    for idx, row in df_assistant.iterrows(): # df_assistant変数（データフレーム）の各行について、インデックス（行番号）をidx変数に、行データをrow変数に代入しながら繰り返し処理を行います。forというキーワードを使って、繰り返し処理を定義します。iterrowsメソッドとは、データフレームの各行に対してインデックスと行データのペアを返すためのものです。メソッドとは、クラスやオブジェクトに属する関数のことです。クラスとは、データや処理などをまとめた設計図のようなものです。オブジェクトとは、クラスから作られた実体のことです。
        input = df_prompter.loc[row.parent_id] # df_prompter変数（データフレーム）からrow変数（行データ）のparent_id列の値に対応する行データを取り出し、inputという変数に代入します。loc属性とは、インデックスや列名などを使ってデータフレームやシリーズから部分的なデータを取り出すためのものです。
        inputs.append(input.text) # inputs変数（リスト）にinput変数（行データ）のtext列の値を追加します。appendメソッドとは、リストに要素を末尾に追加するためのものです。

    df_assistant["instruction"] = inputs # df_assistant変数（データフレーム）にinstruction列を追加し、inputs変数（リスト）の値を代入します。

    df_assistant = df_assistant[df_assistant.lang=="en"] # df_assistant変数（データフレーム）からlang列が"en"である行だけを抽出し、df_assistant変数に再代入します。

    df_assistant = df_assistant[["instruction", "output"]] # df_assistant変数（データフレーム）からinstruction列とoutput列だけを抽出し、df_assistant変数に再代入します。

    return df_assistant # returnというキーワードを使って、df_assistant変数（データフレーム）を関数の返り値（戻り値）として返します。

df_train = prep_data(train) # prep_data関数にtrain変数（データフレーム）を引数として渡して呼び出し、返り値（データフレーム）をdf_trainという変数に代入します。
df_val = prep_data(val) # prep_data関数にval変数（データフレーム）を引数として渡して呼び出し、返り値（データフレーム）をdf_valという変数に代入します。

pd.concat([df_train, df_val]).reset_index(drop=True).to_csv("data/oasst-data/train_full.csv", index=False) # pd.concat関数にdf_train変数とdf_val変数（データフレーム）のリストを引数として渡して呼び出し、返り値（結合されたデータフレーム）に対してreset_indexメソッドとto_csvメソッドを連続して呼び出します。pd.concat関数とは、複数のデータフレームを縦方向や横方向に結合するための関数です。reset_indexメソッドとは、データフレームのインデックスを振り直すためのメソッドです。drop=Trueというオプションは、元のインデックスを削除するためのものです。to_csvメソッドとは、データフレームをCSVファイルに保存するためのメソッドです。CSVファイルとは、カンマで区切られた値の形式でデータを保存するファイルのことです。"data/oasst-data/train_full.csv"という文字列は、保存するファイル名を指定するためのものです。index=Falseというオプションは、インデックスをファイルに保存しないためのものです。
````

````python
# このコードは、pandasという名前のライブラリをインポートし、val_outputsという名前の変数にoutput/demo_oasst-data/validation_predictions.csvという名前のファイルを読み込むためのものです。
# pandasとは、前述と同じ意味です。
import pandas as pd # pandasをpdという名前でインポートします。これにより、pandasの機能をpdという名前で呼び出せます。
val_outputs = pd.read_csv("output/demo_oasst-data/validation_predictions.csv") # pd.read_csv関数に"output/demo_oasst-data/validation_predictions.csv"という文字列（引数）を渡して呼び出し、返り値（データフレーム）をval_outputsという変数に代入します。read_csv関数とは、CSVファイルを読み込んでデータフレームに変換するための関数です。
````

````python
# このコードは、val_outputsという名前の変数（データフレーム）の41番目から42番目の行について、インデックス（行番号）、行データ、指示文、正解の出力、予測された出力を表示するためのものです。
# val_outputsとは、前述と同じ意味です。
for _, row in val_outputs.iloc[41:42].iterrows(): # val_outputs変数（データフレーム）のiloc属性に[41:42]というスライスを渡して部分的なデータフレームを取り出し、そのデータフレームに対してiterrowsメソッドを呼び出し、返り値（インデックスと行データのペア）について繰り返し処理を行います。forというキーワードを使って、繰り返し処理を定義します。_という変数にインデックスを代入し、rowという変数に行データを代入します。_という変数は、使わない値を代入するための慣習的な名前です。iloc属性とは、整数やスライスなどを使ってデータフレームやシリーズから部分的なデータを取り出すためのものです。スライスとは、[開始:終了:間隔]という形式で範囲や順序を表すものです。iterrowsメソッドとは、前述と同じ意味です。
    print("============") # print関数に"============"という文字列（引数）を渡して呼び出します。これにより、"============"という文字列が画面に表示されます。print関数とは、引数に渡された値や文字列などを画面に表示するための関数です。
    print() # print関数に引数を渡さずに呼び出します。これにより、空白行が画面に表示されます。
    print(row.instruction) # print関数にrow変数（行データ）のinstruction列の値（引数）を渡して呼び出します。これにより、指示文が画面に表示されます。
    print() # print関数に引数を渡さずに呼び出します。これにより、空白行が画面に表示されます。
    print("-----Target Answer-----") # print関数に"-----Target Answer-----"という文字列（引数）を渡して呼び出します。これにより、"-----Target Answer-----"という文字列が画面に表示されます。
    print() # print関数に引数を渡さずに呼び出します。これにより、空白行が画面に表示されます。
    print(row.output) # print関数にrow変数（行データ）のoutput列の値（引数）を渡して呼び出します。これにより、正解の出力が画面に表示されます。
    print() # print関数に引数を渡さずに呼び出します。これにより、空白行が画面に表示されます。
    print("-----Predicted Answer-----") # print関数に"-----Predicted Answer-----"という文字列（引数）を渡して呼び出します。これにより、"-----Predicted Answer-----"という文字列が画面に表示されます。
    print() # print関数に引数を渡さずに呼び出します。これにより、空白行が画面に表示されます。
    print(row.pred_output) # print関数にrow変数（行データ）のpred_output列の値（引数）を渡して呼び出します。これにより、予測された出力が画面に表示されます。
    print() # print関数に引数を渡さずに呼び出します。これにより、空白行が画面に表示されます。
````
